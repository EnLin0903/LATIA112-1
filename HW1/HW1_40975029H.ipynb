```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
fm.fontManager.addfont('Microsoft-JhengHei.ttf')
plt.rc('font', family='Microsoft JhengHei') #讓plt可以顯示中文
#plt.rcParams['font.family'] = 'Microsoft JhengHei'  # 指定 'Microsoft JhengHei' 或其他已安装的中文字体

# 匯入CSV格式的檔案
df = pd.read_csv('111_leave.csv', encoding="utf8")
df
```

```python
#刪除最後三行說明文字
df = df.drop(df.index[-3:])
df['Unnamed: 1'] = df['Unnamed: 1'].fillna(method='ffill')#將校名自動填充
df['Unnamed: 2'] = df['Unnamed: 2'].fillna(method='ffill')#將地區自動填充
df['Unnamed: 1'] = df['Unnamed: 1'].str.replace('ˉˉˉ', '', regex=True)
df
```

```python
# 建立除去前兩行的dataframe
df_slice = df.iloc[2:]
df_slice['總計'] = df_slice['總計'].replace(' -', '0') #把字元' -'替換成'0'
df_slice['總計'] = df_slice['總計'].str.replace(',', '', regex=True).astype(int) #過千的數字會出現逗點，因此把字元','移除，並將「總計」index底下的資料轉換成整數型態
df_slice
```

```python
school_list = list(df['Unnamed: 1'].unique())#收集Unnamed: 1底下所有相異資料
school_list = school_list[2:]#排除前兩筆無意義資料
school_list = [item.rstrip("ˉˉˉ") for item in school_list]#將所有資料後端的無意義字元清除
national_count = sum(name.startswith('國立') for name in school_list)#計算序列裡以「國立」起頭的資料
print(f"本資料集總共收集了{len(set(school_list))}所大專院校，其中含有{national_count}間國立大學")

percentage = national_count/int(len(set(school_list)))
percentage = round(percentage, 2)
other_count = int(len(set(school_list))) - national_count

#建立數據
labels = ['國立大學', '私立大學']
sizes = [national_count,other_count]
colors = ['gray', 'lavender']
explode = (0, 0.1)

#繪製圓餅圖
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # 使圓餅圖為等比例
plt.title('111年國、私立大學佔比')
plt.show()
```

```python
data_S1 = df_slice[df_slice['Unnamed: 3'] == '博士'] #把Unnamed: 3欄位中，值為"博士"的資料建立為 data_S1
total_S1 = sum(data_S1['總計'])
print(f"有{total_S1}個博士生辦休學")

data_S2 = df_slice[df_slice['Unnamed: 3'] == '碩士'] #把Unnamed: 3欄位中，值為"碩士"的資料建立為 data_S2
total_S2 = sum(data_S2['總計'])
print(f"有{total_S2}個碩士生辦休學")

data_S3 = df_slice[df_slice['Unnamed: 3'] == '學士'] #把Unnamed: 3欄位中，值為"學士"的資料建立為 data_S3
total_S3 = sum(data_S3['總計'])
print(f"有{total_S3}個學士生辦休學")


excluded_indexes = ['博士', '碩士', '學士'] #建立含有三種學歷的list
data_S4 = df_slice[df_slice['Unnamed: 3'].notna() & ~df_slice['Unnamed: 3'].isin(excluded_indexes)] #把Unnamed: 3欄位中，其他學歷的資料建立為 data_S4

total_S4 = sum(data_S4['總計'])
print(f"有{total_S4}個其他學生辦休學")

#建立數據
categories = ['博士班', '碩士班', '學士班','其他']
student_counts = [total_S1, total_S2, total_S3, total_S4]
colors = ['skyblue', 'lightcoral', 'teal','red']

# 創建條形圖
plt.bar(categories, student_counts, color=colors)
plt.xlabel('等級別')
plt.ylabel('學生人數')
plt.title('111學年度各等級學生休學人數')

# 標示每個條形的數量
for i in range(len(categories)):
    plt.text(categories[i], student_counts[i]-2000, str(student_counts[i]), ha='center', va='bottom')

plt.show()
```

```python
# 根據校名是否為「國立」開頭，並分組求學生人數和
grouped = df_slice.groupby(df_slice['Unnamed: 1'].str.startswith('國立'))['總計'].sum()
print(f"共有{grouped[0]}個國立大學學生休學")
print(f"共有{grouped[1]}個私立大學學生休學")

#建立數據
labels = ['國立大學', '私立大學']
colors = ['gray', 'lavender']
explode = (0, 0.1)

#繪製圓餅圖
plt.pie(grouped, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # 使圓餅圖為等比例
plt.title('國立大學與非國立大學的休學人數比例')
plt.show()
```

```python
# 按校名分组并计算每个学校的学生人数总和
school_total_students = df_slice.groupby('Unnamed: 1')['總計'].sum()

# 对学校按学生人数降序排序，并选择前三名学校
top_schools = school_total_students.sort_values(ascending=False).head(3)


#建立顏色
colors = ['skyblue', 'lightcoral', 'teal','red']

# 創建條形圖
plt.bar(top_schools.index, top_schools.values, color=colors)
plt.xlabel('學校名稱')
plt.ylabel('學生人數')
plt.title('111學年度最多休學人數前三名之大學')

# 標示每個條形的數量
for i, v in enumerate(top_schools.values):
    plt.text(i, v-200, str(v), ha='center', va='bottom')

plt.show()
```
